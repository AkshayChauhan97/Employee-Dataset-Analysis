```python
#Merging tables to collect insights
emp_dept= pd.merge(df_dept, df_dept_emp, on = "department_id", how = "inner") 
emp_dept
```

```python
title_sal= pd.merge(df_emp_title, df_emp_sal, on = "employee_id", how = "inner") 
title_sal
```

```python
# Find the highest salary in each title
max_salary_per_title = title_sal.groupby('title')['amount'].max().reset_index()
max_salary_per_title
```

```python
emp_dept_sal= pd.merge(emp_dept, df_emp_sal, on = "employee_id", how = "inner") 
emp_dept_sal
```

```python
# Find the highest salary in each department
max_salary_per_department = emp_dept_sal.groupby('dept_name')['amount'].max().reset_index()
max_salary_per_department
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
plt.bar(max_salary_per_department['dept_name'], max_salary_per_department['amount'], color='Orange')
plt.xlabel('Department')
plt.ylabel('Highest Salary')
plt.title('Highest Salary in Each Department')
plt.show()
```

```python
emp_title_sal= pd.merge(emp_dept,title_sal , on = "employee_id", how = "inner")
emp_title_sal
```

```python
columns_to_drop=['department_id','from_date','to_date','from_date_x','to_date_x','from_date_y','to_date_y']
emp_title_sal.drop(columns=columns_to_drop, inplace=True)
```

```python
emp_title_sal
```

```python
emp_title_sal.set_index('employee_id', inplace=True)
```

```python
# Group by 'dept_name' and 'title', then find the maximum 'amount' in each group
result_df = emp_title_sal.groupby(['dept_name', 'title'])['amount'].max().reset_index()

# Display the result
print(result_df)
```

```python
# Create a pie chart using seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(15, 15))

# Use 'dept_name' and 'title' as labels, 'amount' as sizes
plt.pie(result_df['amount'], labels=result_df['dept_name'] + ' - ' + result_df['title'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
```

```python
# Avg Salaries among the departments between different titles
title_sal_avg=pd.read_sql_query(""" Select title, avg(amount) AS avg_salary
                                from employees.title
                                inner join employees.salary on employees.title.employee_id=employees.salary.employee_id
                                group by title;""", connection)
plt.figure(figsize=(10, 5))
sns.barplot(data= title_sal_avg, x= 'title', y='avg_salary') 
plt.show()
```

```python
title_sal= pd.read_sql_query(""" Select title, sum(amount) AS amount
                                from employees.title
                                inner join employees.salary on employees.title.employee_id=employees.salary.employee_id
                                group by title;""", connection)
# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(title_sal['amount'], labels=title_sal['title'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Title-wise salaries')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Highest salaries among the departments of active employees
all_salary = pd.read_sql_query("""select d.dept_name, s.amount from employees.salary s left join employees.department_manager mn
on s.employee_id=mn.employee_id left join employees.department d on d.id=mn.department_id
where date_part('year', s.to_date) =9999 and date_part('year', mn.to_date) =9999""", connection) 
sns.barplot(data= all_salary, x= 'amount', y='dept_name') 
plt.show()
```

```python
dept_sal= pd.read_sql_query(""" Select dept_name, sum(amount) AS amount
                                from employees.department_employee
                                inner join employees.salary on employees.department_employee.employee_id=employees.salary.employee_id
                                inner join employees.department
                                on employees.department.id=employees.department_employee.department_id
                                group by dept_name;""", connection)
# Create a pie chart
plt.figure(figsize=(6, 5))
plt.pie(dept_sal['amount'], labels=dept_sal['dept_name'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Department-wise salaries')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Active managers in each department
active_managers=pd.read_sql_query(""" Select dept_name,count(employee_id) as count
                    from employees.department_manager
                    right join employees.department
                    on employees.department_manager.department_id=employees.department.id
                    where extract(year from department_manager.to_date)=9999
                    group by dept_name;""",connection)
active_managers
```

```python
# Active manager who has the highest salary
man_sal=pd.read_sql_query("""Select dept_name,sum(amount) as amount
                    from employees.department_manager
                    inner join employees.salary
                    on employees.department_manager.employee_id = employees.salary.employee_id
                    inner join employees.department 
                    on employees.department.id = employees.department_manager.department_id
                    where extract(year from department_manager.to_date)=9999
                    group by dept_name""",connection)
plt.figure(figsize=(12, 5))
sns.barplot(data= man_sal, x='dept_name',y= 'amount') 
plt.show()
```
